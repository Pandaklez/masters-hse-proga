Алгоритм должен работать так:

строки с текстом разбиваются на отдельные символы и далее в цикле из N итераций: а) считаются статистики встречаемости по парам символов и б) топ-K частотных пар склеиваются в один символ

````python
import re
from collections import defaultdict


def rewrite_one_pair(pair, text_symbols):
    # Меняет во всем тексте последовательность двух строк из pair на одну склееную
    
    text_symbols_out = []
    
    skip_next = False    
    for i in range(len(text_symbols) - 1):
        if skip_next is True:
            skip_next = False
            continue
        if (text_symbols[i], text_symbols[i + 1]) == pair:
            text_symbols_out.append(''.join(pair))
            # skip the next iteration
            skip_next = True
        else:
            text_symbols_out.append(text_symbols[i])
        
    if (text_symbols[-2], text_symbols[-1]) != pair:      
        text_symbols_out.append(text_symbols[-1])  # костылек
        
    return text_symbols_out
   
    
def merge_and_rewrite(top_k_pairs, text_symbols):
    for pair, _ in top_k_pairs:
        # пары склеиваются в один символ
        text_symbols = rewrite_one_pair(pair, text_symbols)
    return text_symbols


def bpe(text_symbols, N, K):
    for i in range(N):
        pairs = defaultdict(int)

        # считаются статистики встречаемости по парам символов
        for i in range(len(text_symbols) - 1):
            # вместо frequency -- сколько раз встретилась в text_symbols последовательность (i, i+1)
            pairs[text_symbols[i], text_symbols[i + 1]] += 1

        top_k_pairs = sorted(pairs.items(), key=lambda x: x[1], reverse=True)[:K]

        # топ-K частотных пар склеиваются в один символ
        #text_symbols чтоб это был словарь итерационно развивающийся
        text_symbols = merge_and_rewrite(top_k_pairs, text_symbols)

        #print("Top-k frequent pairs ", top_k_pairs)
        print("len(text_symbols) ==== ", len(text_symbols))
    return text_symbols
````

Векторизуйте текст (в мешок слов) аналогично TfidfVectorizer, используя токенизатор и idf статистики (инициализируйте*** пустую матрицу размером (N документов, K слов в словаре) и в цикле по всем документам постепенно заполните ее - токенизируйте документ, рассчитайте TF каждого слова (количество вхождений в документе поделить на общее количество слов в документе), умножьте TF на IDF и, используя индексы слов в словаре, запишите получившееся значение в матрицу)

````python
# tokenize texts and make a set of tokens from texts
all_tokens, tokenized_docs = [], []
for doc in documents[:30000]:
    tokenized_doc = tok_sub.encode(doc).ids
    all_tokens.extend(tokenized_doc)
    tokenized_docs.append(tokenized_doc)
    
vocab = set(all_tokens)
print(len(vocab))
````

````python
import math
from scipy.sparse import lil_matrix
from collections import Counter


def fit_transform(documents):
    ### Learn vocabulary and idf, return document-term matrix.
    # documents: list of texts
    
    all_tokens, tokenized_docs = [], []
    for doc in documents:
        tokenized_doc = tok_sub.encode(doc).ids
        all_tokens.extend(tokenized_doc)
        tokenized_docs.append(tokenized_doc)
    
    vocab = set(all_tokens)
    
    # инициализируем пустую матрицу размером (N документов, K слов в словаре)
    N = len(documents)
    K = len(vocab)
    X = lil_matrix((N, K))
    
    idfs = []
    for i, token in enumerate(list(vocab)):
        print(i)
        doc_freq = 0
        # calculate idf
        for doc in tokenized_docs:
            if token in doc:
                doc_freq += 1
        idf = math.log(N/doc_freq)
        idfs.append(idf)
        # calculate tf
        for j, doc in enumerate(tokenized_docs):
            tf = Counter(doc)[token]/len(doc)
            X[j, i] = tf * idf
    return X
````
